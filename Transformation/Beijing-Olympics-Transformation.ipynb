```python
genders.printSchema()
display(genders)
```

```python
genders = genders.withColumn("Female", col("Female").cast(IntegerType()))\
.withColumn("Male", col("Male").cast(IntegerType()))\
.withColumn("Total", col("Total").cast(IntegerType()))
genders.printSchema()
```

```python
medals.printSchema()
display(medals)
```

```python
medals = medals.withColumn("Rank", col("Rank").cast(IntegerType()))\
.withColumn("Gold", col("Gold").cast(IntegerType()))\
.withColumn("Silver", col("Silver").cast(IntegerType()))\
.withColumn("Bronze", col("Bronze").cast(IntegerType()))\
.withColumn("Total", col("Total").cast(IntegerType()))\
.withColumn("Rank by Total", col("Rank by Total").cast(IntegerType()))
medals.printSchema()
```

```python
medals = medals.withColumnRenamed("NOC", "Country")

medals = medals.replace(
    {
        "People's Republic of China": "China",
        "Chinese Taipei": "Chinese Taipei(Taiwan)",
        "Democratic Republic of Timor-Leste": "Timor-Leste",
        "Hong Kong, China": "Hong Kong",
        "Republic of Korea": "South Korea",
        "Republic of Moldova": "Moldova",
        "Islamic Republic of Iran" : "Iran",
        "ROC" : "Russia(ROC)",
        "United States of America": "USA",
        "Virgin Islands, US": "Virgin Islands"

    },
    subset=["Country"]
)
display(medals)
```
